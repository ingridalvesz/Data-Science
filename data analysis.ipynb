### Comparando as notas, médias, medianas e desvio padrão dos filmes Toy Story e Jumanji

```python
notas_do_toy_story = notas.query("filmesId==1")
notas_do_jumanji = notas.query("filmesId==2")
print(len(notas_do_toy_story), len(notas_do_jumanji))
```

```python
plt.boxplot([notas_do_toy_story.nota, notas_do_jumanji.nota])
```

```python
sns.boxplot(x = "filmesId", y = "nota", data = notas.query("filmesId in (1,2)"))
```

```python
print('Média do Toy Story %.2f' % notas_do_toy_story.nota.mean())
print('Média do Jumanji %.2f' % notas_do_jumanji.nota.mean())
```

```python
print('Mediana do Toy Story %.2f' % notas_do_toy_story.nota.median())
print('Mediana do Jumanji %.2f' % notas_do_jumanji.nota.median())
```

```python
print("Desvio padrão do Jumanji %.2f" % notas_do_jumanji.nota.std(), "Desvio padrão do Toy Story %.2f" % notas_do_toy_story.nota.std())
```

## Data visualization: Uma nova lista de filmes tmdb

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
contagem_de_linguas = tmdb['original_language'].value_counts().to_frame().reset_index()
contagem_de_linguas.columns = ['original_language', 'total']
contagem_de_linguas.head()
```

```python
total_por_linguas = tmdb['original_language'].value_counts()
total_geral = total_por_linguas.sum()
total_de_ingles = total_por_linguas.loc['en']
total_das_outras = total_geral - total_de_ingles
print(total_de_ingles, total_das_outras)
```

```python
dados = {
    'linguas' : ['ingles', 'outros'],
    'total' : [total_de_ingles, total_das_outras]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(x = 'linguas', y = 'total', data = dados)
plt.title('Gráfico das outras línguas em comparação com o inglês')
```

```python
plt.pie(dados['total'], labels = dados['linguas'])
plt.title('Gráfico das outras línguas em comparação com o inglês')
```

```python
total_filmes_sem_linguas_en = tmdb.query("original_language != 'en'").original_language.value_counts()
```

```python
filmes_sem_linguas_en = tmdb.query("original_language != 'en'")

sns.catplot(x = 'original_language', kind = 'count',
             data = filmes_sem_linguas_en,
             aspect = 3,
             palette = 'dark:salmon_r',
             order = total_filmes_sem_linguas_en.index )

plt.title('Gráfico de todas as línguas com exeção da língua inglesa')
```

## Utilizando exemplos com numpy na criação de médias e medianas 

```python
filme1 = np.append(np.array([2.5] * 10), np.array([3.5] * 10))
filme2 = np.append(np.array([5] * 10), np.array([1] * 10))

print(filme1.mean(), filme2.mean())
```

```python
print(np.median(filme1), np.median(filme2))
```

```python
plt.boxplot([filme1,filme2])
```
