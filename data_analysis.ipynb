## Análise de dados dos filmes

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

notas = pd.read_csv('ratings.csv')
notas.columns = ['usuarioId', 'filmesId', 'nota', 'momento']
filmes = pd.read_csv('movies.csv')
filmes.columns = ['filmesId','titulosId','generos']
filmes.head()
```

```python
notas.head()
```

```python
filmes.head()
```

#Análise sobre notas de filmes

```python
notas.query('filmesId==1').nota.mean()
```

# Média dos filmes

```python
medias_por_filmes = notas.groupby('filmesId').mean().nota
medias_por_filmes.head()
```

```python
medias_por_filmes.plot(kind = 'hist')
plt.title('Histograma das médias dos filmes')
```

```python
sns.distplot(medias_por_filmes)
plt.title('Histograma das médias dos filmes')
```

```python
sns.boxplot(medias_por_filmes)
plt.title('Histograma das médias dos filmes')
```

```python
medias_por_filmes.describe()
```

# Mediana do filmes

```python
medianas_por_filmes = notas.groupby('filmesId').median().nota
medianas_por_filmes.head()
```

```python
medianas_por_filmes.plot(kind = 'hist')
plt.title('Histograma das medianas dos filmes')
```

```python
sns.distplot(medianas_por_filmes)
plt.title('Histograma das medianas dos filmes')
```

```python
sns.boxplot(medianas_por_filmes)
plt.title('Histograma das medianas dos filmes')
```
