Para uma melhor visualização dos dados iremos adicionar novas colunas,  uma delas deve conter os gastos mensais de cada imóvel, incluindo aluguel e do condomínio

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url, sep = ';')
dados.head()
```

```python
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
```

Em outra coluna deve conter os gastos anuais por imóvel, ou seja, IPTU mais 12 meses de aluguel e condomínio.

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
```

A próxima coluna deve possuir a descrição das principais informações dos imóveis que serão apresentadas no site:
tipo de imóvel, bairro, quantidade de quartos e vagas de garagem

###### Utilizamos .astype(str) para todos pertencem a mesma classe

```python
dados['Discricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                    dados['Quartos'].astype(str) + ' quarto(s) ' + \
                    ' e ' + dados['Vagas'].astype(str) + ' vasga(s) de garagem '
dados.head()
```

Essa nova coluna deve informar apenas imóvel que possuem ou não suítes, sem se importar com a quantidade.

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não") 
dados.head()
```
